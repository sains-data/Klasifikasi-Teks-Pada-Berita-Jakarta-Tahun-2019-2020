```python
kategori = pd.get_dummies(df.tonality)
df = pd.concat([df, kategori], axis=1)
df = df.drop(columns='tonality')
df.sample(5)
```

```python
feature = df['isu']
label = df[['NEGATIF', 'NETRAL', 'POSITIF']].values
```

```python
from sklearn.model_selection import train_test_split
feature_latih, feature_test, label_latih, label_test = train_test_split(feature, label, test_size=0.2, random_state=42)
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences


pad_type = 'pre'
trunc_type = 'pre'

# Tokenize our training data
tokenizer = Tokenizer(num_words=2000, oov_token='x')
tokenizer.fit_on_texts(feature_latih)
tokenizer.fit_on_texts(feature_test)

# Encode training data sentences into sequences
sekuens_latih = tokenizer.texts_to_sequences(feature_latih)
sekuens_test = tokenizer.texts_to_sequences(feature_test)

# Get max training sequence length
maxlen = max([len(x) for x in sekuens_latih])

# Pad the training sequences
padded_latih = pad_sequences(sekuens_latih, padding=pad_type, truncating=trunc_type, maxlen=maxlen)
padded_test = pad_sequences(sekuens_test, padding=pad_type, truncating=trunc_type, maxlen=maxlen)
```

```python
model = tf.keras.Sequential([
  tf.keras.layers.Embedding(input_dim=2000, output_dim=8),
  tf.keras.layers.Dropout(0.5),
  tf.keras.layers.LSTM(32),
  tf.keras.layers.Dropout(0.5),
  tf.keras.layers.Dense(32, activation='sigmoid'),
  tf.keras.layers.Dropout(0.5),
  tf.keras.layers.Dense(3, activation='sigmoid')
  ])
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
class myCallback(tf.keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs={}):
    if(logs.get('accuracy')>0.9 and logs.get('val_accuracy')>0.9):
      print("\nAkurasi telah mencapai >90%!")
      self.model.stop_training = True
callbacks = myCallback()
```

```python
num_epochs = 40
hist = model.fit(padded_latih, label_latih, epochs=num_epochs,
                    validation_data=(padded_test, label_test),
                    verbose=2, callbacks=[callbacks])
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

# Prediksi label dari data uji
label_prediksi = model.predict(padded_test)
label_prediksi = [list(prediksi).index(max(prediksi)) for prediksi in label_prediksi]

# Konversi one-hot encoded labels menjadi label kelas tunggal
label_asli = [list(label).index(1) for label in label_test]

# Buat confusion matrix
cm = confusion_matrix(label_asli, label_prediksi)

# Visualisasi confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['NEGATIF', 'NETRAL', 'POSITIF'], yticklabels=['NEGATIF', 'NETRAL', 'POSITIF'])
plt.title('Confusion Matrix')
plt.show()
```

```python
# summarize history for accuracy
plt.plot(hist.history['accuracy'])
plt.plot(hist.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
# summarize history for loss
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper right')
plt.show()
```

```python
# Memuat model dan tokenizer
import streamlit as st
import pandas as pd
import numpy as np
from tensorflow.keras.models import load_model
from transformers import AutoTokenizer
import re
import string

model_path = "/content/my_model.h5"

max_text_length = 400  # sesuaikan dengan settingan Anda

# Load model
model = load_model(model_path)

# Load tokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

# Fungsi pembersihan teks
def preprocess(text):
    new_text = []
    for t in text.split(" "):
        t = "@user" if t.startswith("@") and len(t) > 1 else t
        t = "http" if t.startswith("http") else t
        new_text.append(t)
    return " ".join(new_text)

# Process the input and return prediction
def sentiment_analysis(text):
    text = preprocess(text)

    # Tokenize input text
    encoded_input = tokenizer(text, return_tensors="tf", max_length=max_text_length, truncation=True)

    # Perform prediction
    output = model.predict(encoded_input)
    scores_ = output[0][0]

    # Format output dict of scores
    labels = ["Negative", "Neutral", "Positive"]
    scores = {l: float(s) for (l, s) in zip(labels, scores_)}

    return scores

# UI Streamlit
st.title("Aplikasi Prediksi Komentar Beracun")
teks_komentar = st.text_area("Masukkan Komentar:")
file_excel = st.file_uploader("Atau unggah file dengan komentar (Excel atau CSV):", type=["xlsx", "csv"])

if st.button("Prediksi"):
    if teks_komentar:
        prediction = sentiment_analysis(teks_komentar)
        st.write("Hasil Prediksi:")
        st.write(prediction)
    elif file_excel:
        df = pd.read_excel(file_excel) if file_excel.type == "application/vnd.openxmlformats-officedocument.spreadsheetml.sheet" else pd.read_csv(file_excel)
        predictions = df["Komentar"].apply(sentiment_analysis)
        st.write("Hasil Prediksi:")
        st.write(predictions)
    else:
        st.warning("Masukkan teks komentar atau unggah file terlebih dahulu.")
```
